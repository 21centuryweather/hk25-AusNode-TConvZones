```python
from statsmodels.tsa.seasonal import STL
from statsmodels.tsa.tsatools import detrend
from scipy import stats
import xarray as xr
import numpy as np

# detrending and anomaly calculation


def detrend1d_check(arr, period):
    # if np.nansum(arr) != 0:
    #     trend = mk.hamed_rao_modification_test(arr).trend
    # else:
    #     trend = 'no trend'
    # if trend != 'no trend':
    #     res = STL(arr, period = period).fit()
    #     arr_det = arr - res.trend
    #     return arr_det
    # else:
    #     return arr
    arr = arr.copy()  # to counter buffer is read-only error
    res = STL(arr, period=period).fit()
    arr_det = arr - res.trend
    # arr_det = detrend(arr, order=2)
    return arr_det


def detrend_separate_check(da, dim, period):
    # da = da.copy()
    return xr.apply_ufunc(
        detrend1d_check,
        da,
        input_core_dims=[[dim]],
        output_core_dims=[[dim]],
        kwargs=dict(period=period),
        vectorize=True,
        dask="parallelized",
    )
```

```python
nino34_ts_index = detrend_separate_check(nino34_ts, 'time', period=15)
```

```python
# plt.figure(figsize=(12,4))
plt.plot(nino34_ts_index.time.sel(time = slice('2000', '2014')).dt.year, nino34_ts_index.sel(time = slice('2000', '2014')))
plt.axhline(0.0)
plt.ylim(-1, 1)
```
